```python
import pandas as pd; #used for csv files
import seaborn as sb; #used for plots..
import matplotlib.pyplot as plt; #This is also used for plots..
import numpy as np; #This is used for multi dimentional arrays... numpy..

#importing the dataset into a variable..
#The names is will add as the header dataframe .. 
haberman = pd.read_csv( '../Downloads/haberman.csv',names = ["Age","Op_Year","axil_nodes_det","Surv_status"]);
```

```python
haberman.plot(kind = "scatter" , x= "axil_nodes_det" , y = "Surv_status")
plt.show()
```

```python
#lets try pair plot...
plt.close()
sb.set_style("whitegrid")
sb.pairplot(haberman,hue="Surv_status",height=3)
plt.show()
```

```python
#Lets try the 2D scatter plot if we can understand any thing from it..
plt.close()
sb.FacetGrid(haberman,hue="Surv_status",size=6) \
    .map(plt.scatter,"Age","axil_nodes_det") \
    .add_legend()
plt.show()
```

## Observations from the above...2D scatter plot...
 1. There are lot of Dead cases between 40 and 70 age..
 2. If you observe the percentage of the people lived if the axil_node_det in between 0 to 10 is more than that of the percentage of the people lived after 10...
     2.1 Females with more than 10 axil_node_det... more cases of less survival...

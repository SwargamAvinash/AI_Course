```python
#importing all the libraries that I have needed in this..
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt 
import numpy as np


iris = pd.read_csv('../Downloads/Iris.csv')
```

```python
#For a normal x-axis and y-axis plot... var.plot... the arguments change will change the thing..!!
iris.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm')
plt.show()#Fishy need to know about this...
```

```python
sb.set_style('whitegrid')
sb.FacetGrid(iris,hue='Species',height=5).map(plt.scatter,"SepalLengthCm","SepalWidthCm").add_legend()
plt.show()
```

```python
iris_setosa = iris.loc[iris["Species"] == "Iris-setosa"]
#print(iris_setosa)
iris_virginica = iris.loc[iris["Species"] == "Iris-virginica"]
iris_versicolor = iris.loc[iris["Species"] == "Iris-versicolor"]
#print(iris_versicolor)
```

```python
sb.FacetGrid(iris,hue="Species",height=7) \
    .map(sb.distplot,"PetalLengthCm") \
    .add_legend()
plt.show()
```

```python
sb.violinplot(x="Species",y="PetalLengthCm",data=iris,height=10)
plt.show()
```
